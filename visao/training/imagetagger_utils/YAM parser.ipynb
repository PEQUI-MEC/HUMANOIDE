```python
def yaml_to_list(path):
    annotation_path = os.path.join(os.getcwd(), (path))
    lista =[]
    count = []
    for filepath in glob.glob(annotation_path + '/*.txt'):
        stream = open(filepath, 'r') 
        yaml = load(stream)
        for d, dataset in yaml.items():
            for i, imagem in dataset.items(): 
                if(i not in ['description', 'team', 'location']):
                    label = []
                    for annotation in imagem['annotations']:
                        x = []
                        if(annotation['label'] in ['ball', 'goalpost', 'robot']):
                            if(annotation['present']):
                                x = (filepath.split('/')[-1][:-4]+'/'+i,
                                            imagem['dimensions'][0],
                                            imagem['dimensions'][1],
                                            annotation['label'],
                                            annotation['bbox'][0],
                                            annotation['bbox'][1],
                                            annotation['bbox'][2],
                                            annotation['bbox'][3])
                                lista.append(x)
                            else:
                                x = (filepath.split('/')[-1][:-4]+'/'+i,
                                            imagem['dimensions'][0],
                                            imagem['dimensions'][1],
                                            annotation['label'],
                                            None,
                                            None,
                                            None,
                                            None)
                                lista.append(x)
                            label.append(annotation['label'])
                    count.append([filepath.split('/')[-1][:-4], label])
        stream.close()
    return lista,count

#column_name = ['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
#pbfile = pd.DataFrame(list, columns=column_name)
#pbfile.to_csv(('_labels.csv'), index=None)
```
